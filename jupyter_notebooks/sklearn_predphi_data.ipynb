```python
import pandas as pd
import os
```

```python
data_folder = '/home/matt/Documents/projectx/PredPHI/data/' # in PredPHI repo https://github.com/xialab-ahu/PredPHI
```

```python
train_X_folder = 'trainingfeatures/'
train_Y_name = 'training_set.csv'
test_X_folder = 'testfeatures/'
test_Y_name = 'test_set.csv'
test_Y_name_alternate = 'test_random.csv'
# also try 'train and test kmeans files -- don't have labels tho ??

#training_labels = pd.read_csv('mediumdata/testseq_AAC.csv', sep='\t', index_col=0)
#aac.shape
```

```python
# load all data files into data frames
train_X = pd.DataFrame()
train_Y = pd.read_csv(data_folder + train_Y_name, sep = '\t')
test_X = pd.DataFrame()
test_Y = pd.read_csv(data_folder + test_Y_name, sep = '\t')
```

```python
train_Y.index = train_Y['phage'] + ',' + train_Y['host']
test_Y.index = test_Y['phage'] + ',' + test_Y['host']
```

```python
train_Y
```

```python
train_X_data = []
for f in os.listdir(data_folder + train_X_folder):
    data = pd.read_csv(data_folder + train_X_folder + f, header = None).T
    data.index = [f[:-4]]
    train_X_data.append(data)
train_X = train_X.append(train_X_data)
```

```python
test_X_data = []
for f in os.listdir(data_folder + test_X_folder):
    data = pd.read_csv(data_folder + test_X_folder + f, header = None).T
    data.index = [f[:-4]]
    test_X_data.append(data)
test_X = test_X.append(test_X_data)
```

```python
train_X
```

```python
test_X
```

```python
train_Y = train_Y.loc[train_X.index]
test_Y = test_Y.loc[test_X.index]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression(random_state=0, max_iter=1000).fit(train_X, train_Y['class'])
```

```python
print(lr.score(train_X, train_Y['class']))
print(lr.score(test_X, test_Y['class']))
```

```python
svm = make_pipeline(StandardScaler(), SVC(gamma='auto')).fit(train_X, train_Y['class'])
```

```python
print(svm.score(train_X, train_Y['class']))
print(svm.score(test_X, test_Y['class']))
```

```python
rf = RandomForestClassifier(random_state=0, n_estimators = 500).fit(train_X, train_Y['class'])
```

```python
print(rf.score(train_X, train_Y['class']))
print(rf.score(test_X, test_Y['class']))
```

```python
rf = RandomForestClassifier(random_state=0, n_estimators = 1000).fit(train_X, train_Y['class'])
print(rf.score(train_X, train_Y['class']))
print(rf.score(test_X, test_Y['class']))
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
names = ["Nearest Neighbors", "Linear SVM", "RBF SVM", "Gaussian Process",
         "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
         "Naive Bayes", "QDA"]

classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis()]
```

```python
for name, clf in zip(names, classifiers):
    clf.fit(train_X, train_Y['class'])
    print(name)
    print(clf.score(train_X, train_Y['class']))
    print(clf.score(test_X, test_Y['class']))
```
